```python
# =========================================================
#  FEATURE ENGINEERING
# =========================================================

# Convert datetime columns
train_df['trans_date_trans_time'] = pd.to_datetime(train_df['trans_date_trans_time'])
test_df['trans_date_trans_time'] = pd.to_datetime(test_df['trans_date_trans_time'])

train_df['dob'] = pd.to_datetime(train_df['dob'])
test_df['dob'] = pd.to_datetime(test_df['dob'])

# =========================================================
# AGE FEATURE
# =========================================================

train_df['age'] = train_df['trans_date_trans_time'].dt.year - train_df['dob'].dt.year
test_df['age'] = test_df['trans_date_trans_time'].dt.year - test_df['dob'].dt.year

# =========================================================
# DISTANCE FEATURE
# =========================================================

def haversine(lat1, lon1, lat2, lon2):
    R = 6371

    lat1 = np.radians(lat1)
    lon1 = np.radians(lon1)
    lat2 = np.radians(lat2)
    lon2 = np.radians(lon2)

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = (np.sin(dlat / 2) ** 2 +
         np.cos(lat1) * np.cos(lat2) *
         np.sin(dlon / 2) ** 2)

    c = 2 * np.arcsin(np.sqrt(a))

    return R * c

train_df['distance'] = haversine(
    train_df['lat'],
    train_df['long'],
    train_df['merch_lat'],
    train_df['merch_long']
)

test_df['distance'] = haversine(
    test_df['lat'],
    test_df['long'],
    test_df['merch_lat'],
    test_df['merch_long']
)

# =========================================================
# TIME FEATURES
# =========================================================

train_df['hour'] = train_df['trans_date_trans_time'].dt.hour
train_df['day'] = train_df['trans_date_trans_time'].dt.day
train_df['month'] = train_df['trans_date_trans_time'].dt.month

test_df['hour'] = test_df['trans_date_trans_time'].dt.hour
test_df['day'] = test_df['trans_date_trans_time'].dt.day
test_df['month'] = test_df['trans_date_trans_time'].dt.month
```

```python
# =========================================================
# LABEL ENCODING (FINAL FIX)
# =========================================================

from sklearn.preprocessing import LabelEncoder

categorical_cols = [
    'merchant',
    'category',
    'gender',
    'city',
    'state',
    'job'
]

encoders = {}

for col in categorical_cols:

    train_df[col] = train_df[col].astype(str)
    test_df[col] = test_df[col].astype(str)

    le = LabelEncoder()

    combined_data = pd.concat(
        [train_df[col], test_df[col]],
        axis=0
    )

    le.fit(combined_data)

    train_df[col] = le.transform(train_df[col])
    test_df[col] = le.transform(test_df[col])

    encoders[col] = le
```

```python
# =========================================================
#  DROP USELESS COLUMNS
# =========================================================

drop_cols = [
    'Unnamed: 0',
    'first',
    'last',
    'street',
    'trans_num',
    'dob',
    'trans_date_trans_time'
]

train_df.drop(columns=drop_cols, inplace=True)
test_df.drop(columns=drop_cols, inplace=True)
```

```python
# =========================================================
#  FEATURES & TARGET
# =========================================================

X_train_full = train_df.drop('is_fraud', axis=1)
y_train_full = train_df['is_fraud']

X_test = test_df.drop('is_fraud', axis=1)
y_test = test_df['is_fraud']
```

```python
# =========================================================
#  TRAIN / VALIDATION SPLIT
# =========================================================

X_train, X_val, y_train, y_val = train_test_split(
    X_train_full,
    y_train_full,
    test_size=0.20,
    stratify=y_train_full,
    random_state=42
)

print("\nTrain Shape:", X_train.shape)
print("Validation Shape:", X_val.shape)
print("Test Shape:", X_test.shape)
```

```python
# =========================================================
#  FEATURE SCALING
# =========================================================

scaler = RobustScaler()

cols_to_scale = [
    'amt',
    'city_pop',
    'age',
    'distance',
    'hour',
    'day',
    'month'
]

X_train_scaled = X_train.copy()
X_val_scaled = X_val.copy()
X_test_scaled = X_test.copy()

X_train_scaled[cols_to_scale] = scaler.fit_transform(
    X_train[cols_to_scale]
)

X_val_scaled[cols_to_scale] = scaler.transform(
    X_val[cols_to_scale]
)

X_test_scaled[cols_to_scale] = scaler.transform(
    X_test[cols_to_scale]
)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, precision_recall_curve

def evaluate_model(name, model, X_test, y_test, threshold=0.5):
    print(f"\n================ {name} ================\n")
    
    # Probabilities
    y_prob = model.predict_proba(X_test)[:, 1]
    
    # Threshold tuning
    y_pred = (y_prob >= threshold).astype(int)
    
    # Metrics
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
    
    print("ROC-AUC:", roc_auc_score(y_test, y_prob))
```

```python
evaluate_model("Random Forest", base_model, X_test, y_test, threshold=0.2)
```

```python
evaluate_model("XGBoost", xgb_model, X_test, y_test, threshold=0.2)
```

```python
results = {}

def store_result(name, model, X_test, y_test):
    y_prob = model.predict_proba(X_test)[:, 1]
    results[name] = roc_auc_score(y_test, y_prob)

store_result("Random Forest", base_model, X_test, y_test)
store_result("XGBoost", xgb_model, X_test, y_test)
store_result("LightGBM", lgbm_model, X_test, y_test)
store_result("Logistic Regression", log_model, X_test_scaled, y_test)

print("\n===== FINAL MODEL COMPARISON (ROC-AUC) =====")
for k, v in sorted(results.items(), key=lambda x: x[1], reverse=True):
    print(f"{k}: {v:.4f}")
```
