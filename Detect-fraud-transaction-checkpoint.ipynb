```python
#
initial_drop = [
    'Unnamed: 0', 'first', 'last', 'street', 
    'city', 'state', 'zip', 'trans_num', 'unix_time'
]

train_data = train_data.drop(columns=initial_drop)
val_data = val_data.drop(columns=initial_drop)
test_df = test_df.drop(columns=initial_drop)

# 2. The Velocity step (must be done while the cc_num and time are still present)
def add_velocity_feature(df):
    # 1. Ensure the time format and order are correct.
    df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
    df = df.sort_values(by=['cc_num', 'trans_date_trans_time'])
    
    # 2. Calculating the number of transactions in the last 24 hours

# We set the time as the index so the scrolling window correctly understands the '24h' window.
    df = df.set_index('trans_date_trans_time')
    
    df['trans_velocity_24h'] = (
        df.groupby('cc_num')['amt'] # We choose any column, like amt, just to count it
        .rolling('24h')
        .count()
        .reset_index(level=0, drop=True) # We remove the extra cc_num that appeared in the index
    )
    
    # 3. We revert the time to a regular column and restore the original index.
    df = df.reset_index()
    return df


train_data = add_velocity_feature(train_data)
val_data = add_velocity_feature(val_data)
test_df = add_velocity_feature(test_df)



# 3. Engineering the remaining features (age, distance, hour) and final scan
def engineer_final_features(df):
    df['dob'] = pd.to_datetime(df['dob'])
    

    df['trans_hour'] = df['trans_date_trans_time'].dt.hour
    df['trans_day_of_week'] = df['trans_date_trans_time'].dt.dayofweek
    df['age'] = (df['trans_date_trans_time'].dt.year - df['dob'].dt.year)
    df['distance'] = np.sqrt(
        (df['lat'] - df['merch_lat'])**2 + 
        (df['long'] - df['merch_long'])**2
    )
    
  
    cols_to_remove = ['trans_date_trans_time', 'dob', 'cc_num']
    return df.drop(columns=cols_to_remove)


train_data = engineer_final_features(train_data)
val_data = engineer_final_features(val_data)
test_df = engineer_final_features(test_df)
```

#Encoding

```python
from sklearn.model_selection import StratifiedKFold

# =========================================================
# KFold Target Encoding (Leakage-Free)
# =========================================================

target_cols = ['category', 'job', 'merchant']

# عدد الـ folds
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# global mean
global_mean = train_data['is_fraud'].mean()

for col in target_cols:

    # column جديدة مؤقتة
    train_data[col + '_enc'] = np.nan

    # KFold Encoding للـ Train
    for train_idx, valid_idx in kf.split(train_data, train_data['is_fraud']):

        X_train_fold = train_data.iloc[train_idx]
        X_valid_fold = train_data.iloc[valid_idx]

        # حساب المتوسط داخل الـ fold فقط
        fold_means = X_train_fold.groupby(col)['is_fraud'].mean()

        # تطبيقه على الـ validation fold
        train_data.iloc[valid_idx,
                        train_data.columns.get_loc(col + '_enc')] = (
            X_valid_fold[col].map(fold_means)
        )

    # ملء القيم الفارغة
    train_data[col + '_enc'] = train_data[col + '_enc'].fillna(global_mean)

    # حساب encoding كامل للـ validation/test
    full_means = train_data.groupby(col)['is_fraud'].mean()

    val_data[col + '_enc'] = val_data[col].map(full_means)
    test_df[col + '_enc'] = test_df[col].map(full_means)

    val_data[col + '_enc'] = val_data[col + '_enc'].fillna(global_mean)
    test_df[col + '_enc'] = test_df[col + '_enc'].fillna(global_mean)

# حذف الأعمدة الأصلية
train_data.drop(columns=target_cols, inplace=True)
val_data.drop(columns=target_cols, inplace=True)
test_df.drop(columns=target_cols, inplace=True)

# 2. Gender  0 , 1
gender_map = {'M': 1, 'F': 0}
train_data['gender'] = train_data['gender'].map(gender_map)
val_data['gender'] = val_data['gender'].map(gender_map)
test_df['gender'] = test_df['gender'].map(gender_map)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, precision_recall_curve

def evaluate_model(name, model, X_test, y_test, threshold=0.5):
    print(f"\n================ {name} ================\n")
    
    # Probabilities
    y_prob = model.predict_proba(X_test)[:, 1]
    
    # Threshold tuning
    y_pred = (y_prob >= threshold).astype(int)
    
    # Metrics
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
    
    print("ROC-AUC:", roc_auc_score(y_test, y_prob))
```

```python
evaluate_model("Random Forest", base_model, X_test, y_test, threshold=0.2)
```

```python
evaluate_model("XGBoost", xgb_model, X_test, y_test, threshold=0.2)
```

```python
results = {}

def store_result(name, model, X_test, y_test):
    y_prob = model.predict_proba(X_test)[:, 1]
    results[name] = roc_auc_score(y_test, y_prob)

store_result("Random Forest", base_model, X_test, y_test)
store_result("XGBoost", xgb_model, X_test, y_test)
store_result("LightGBM", lgbm_model, X_test, y_test)
store_result("Logistic Regression", log_model, X_test_scaled, y_test)

print("\n===== FINAL MODEL COMPARISON (ROC-AUC) =====")
for k, v in sorted(results.items(), key=lambda x: x[1], reverse=True):
    print(f"{k}: {v:.4f}")
```
